#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
demo_dummie = pd.get_dummies(demo, columns = ["sexo", "posse_de_veiculo", "posse_de_imovel", "tipo_renda", "educacao", "estado_civil", "tipo_residencia"])
demo_dummie
````

````python
demo_dummie.info()
````

````python
def add_underscore(df):
    new_columns = [col.replace(" ", "_") for col in df.columns]
    df.columns = new_columns
    return df

demo_dummie = add_underscore(demo_dummie.copy())
````

````python
demo_dummie.info()
````

````python
demo_dummie = demo_dummie.rename(columns={"estado_civil_Single_/_not_married": "estado_civil_Single_not_married"})
demo_dummie = demo_dummie.rename(columns={"tipo_residencia_House_/_apartment": "tipo_residencia_House_apartment"})
demo_dummie = demo_dummie.rename(columns={"tipo_residencia_Co-op_apartment": "tipo_residencia_Co_op_apartment"})
````

````python
demo_dummie.mau = demo_dummie.mau.astype(int)
demo_dummie.sexo_F = demo_dummie.sexo_F.astype(int)
demo_dummie.sexo_M = demo_dummie.sexo_M.astype(int)
demo_dummie.posse_de_veiculo_N = demo_dummie.posse_de_veiculo_N.astype(int)
demo_dummie.posse_de_veiculo_Y = demo_dummie.posse_de_imovel_Y.astype(int)
demo_dummie.tipo_renda_Commercial_associate = demo_dummie.tipo_renda_Commercial_associate.astype(int)
demo_dummie.tipo_renda_Pensioner = demo_dummie.tipo_renda_Pensioner.astype(int)
demo_dummie.tipo_renda_State_servant = demo_dummie.tipo_renda_State_servant.astype(int)
demo_dummie.tipo_renda_Student = demo_dummie.tipo_renda_Student.astype(int)
demo_dummie.tipo_renda_Working = demo_dummie.tipo_renda_Working.astype(int)
demo_dummie.educacao_Academic_degree = demo_dummie.educacao_Academic_degree.astype(int)
demo_dummie.educacao_Higher_education = demo_dummie.educacao_Higher_education.astype(int)
demo_dummie.educacao_Incomplete_higher = demo_dummie.educacao_Incomplete_higher.astype(int)
demo_dummie.educacao_Lower_secondary = demo_dummie.educacao_Lower_secondary.astype(int)
demo_dummie.educacao_Secondary = demo_dummie.educacao_Secondary.astype(int)
demo_dummie.estado_civil_Civil_marriage = demo_dummie.estado_civil_Civil_marriage.astype(int)
demo_dummie.estado_civil_Married = demo_dummie.estado_civil_Married.astype(int)
demo_dummie.estado_civil_Separated = demo_dummie.estado_civil_Separated.astype(int)
demo_dummie.estado_civil_Single_not_married = demo_dummie.estado_civil_Single_not_married.astype(int)
demo_dummie.estado_civil_Widow = demo_dummie.estado_civil_Widow.astype(int)
demo_dummie.tipo_residencia_Co_op_apartment = demo_dummie.tipo_residencia_Co_op_apartment.astype(int) 
demo_dummie.tipo_residencia_House_apartment = demo_dummie.tipo_residencia_House_apartment.astype(int)
demo_dummie.tipo_residencia_Municipal_apartment = demo_dummie.tipo_residencia_Municipal_apartment.astype(int)
demo_dummie.tipo_residencia_Office_apartment = demo_dummie.tipo_residencia_Office_apartment.astype(int)
demo_dummie.tipo_residencia_Rented_apartment = demo_dummie.tipo_residencia_Rented_apartment.astype(int)
demo_dummie.tipo_residencia_With_parents = demo_dummie.tipo_residencia_With_parents.astype(int)

````
